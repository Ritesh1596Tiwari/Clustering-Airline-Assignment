```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

sns.set_style('darkgrid')
```

```python
df = pd.read_csv('EastWestAirlines_1.csv')
```

```python
df
```

```python
df.head()
```

```python
df1 = df.drop(columns=['ID#','Award?'])
```

```python
df1
```

```python
df1.describe()
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

x = sc.fit_transform(df1)

x
```

```python
from sklearn.cluster import KMeans

km = KMeans()

y = km.fit_predict(x)

y
```

```python
km.cluster_centers_
```

```python
x = pd.DataFrame(x)
x['Cluster'] = y

x
```

```python
#to find optimal value we use ELBOW GRAPH
```

```python
x1 = x.iloc[:,:-1]

x1
```

```python
wcss = []

for i in range(1,20):
    km= KMeans(n_clusters=i)
    km.fit_predict(x1)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,20),wcss)
plt.title('Elbow Graph')
plt.xlabel('K-Value')
plt.ylabel('WCSS')
```
